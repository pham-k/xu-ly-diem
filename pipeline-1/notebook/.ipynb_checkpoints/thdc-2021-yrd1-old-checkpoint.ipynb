```python
# const
PATH_RCC1 = '../data/2021/response/thdc-2021-yrd1-cc1 (Responses).xlsx'
PATH_RCC2 = '../data/2021/response/thdc-2021-yrd1-cc2 (Responses).xlsx'
PATH_RCC3 = '../data/2021/response/thdc-2021-yrd1-cc3 (Responses).xlsx'
PATH_RQT1 = '../data/2021/response/thdc-2021-yrd1-qt1 (Responses).xlsx'
PATH_RTHILT1 = '../data/2021/response/thdc-2021-yrd1-thilt1 (Responses).xlsx'
PATH_RTHILT2 = '../data/2021/response/thdc-2021-yrd1-thilt2 (Responses).xlsx'
PATH_RTHITH1 = '../data/2021/response/thdc-2021-yrd1-thith1 (Responses).xlsx'
PATH_RTHITH2 = '../data/2021/response/thdc-2021-yrd1-thith2 (Responses).xlsx'

PATH_CUTPOINT = '../data/2021/reference/cutpoint.csv'
PATH_ATTENDEE11 = '../data/2021/mau-import-diem-import/mau-import-diem-thdc-2021-yrd1-y-2021-lan1.xlsx'
PATH_ATTENDEE12 = '../data/2021/mau-import-diem-import/mau-import-diem-thdc-2021-yrd1-y-2021-lan2.xlsx'
PATH_ATTENDEE21 = '../data/2021/mau-import-diem-import/mau-import-diem-thdc-2021-yrd1-rhm-2021-lan1.xlsx'
PATH_ATTENDEE22 = '../data/2021/mau-import-diem-import/mau-import-diem-thdc-2021-yrd1-rhm-2021-lan2.xlsx'
PATH_ATTENDEE31 = '../data/2021/mau-import-diem-import/mau-import-diem-thdc-2021-yrd1-duoc-2021-lan1.xlsx'
PATH_ATTENDEE32 = '../data/2021/mau-import-diem-import/mau-import-diem-thdc-2021-yrd1-duoc-2021-lan2.xlsx'

cutpoint = pd.read_csv(PATH_CUTPOINT, index_col=None)
# y-2021
attendee11 = util.get_attendee(PATH_ATTENDEE11, course_id='thdc-2021-yrd1', class_id='y-2021')
attendee12 = util.get_attendee(PATH_ATTENDEE12, course_id='thdc-2021-yrd1', class_id='y-2021')
# rhm-2021
attendee21 = util.get_attendee(PATH_ATTENDEE21, course_id='thdc-2021-yrd1', class_id='rhm-2021')
attendee22 = util.get_attendee(PATH_ATTENDEE22, course_id='thdc-2021-yrd1', class_id='rhm-2021')
# duoc-2021
attendee31 = util.get_attendee(PATH_ATTENDEE31, course_id='thdc-2021-yrd1', class_id='duoc-2021')
attendee32 = util.get_attendee(PATH_ATTENDEE32, course_id='thdc-2021-yrd1', class_id='duoc-2021')

exam = util.parse_exam('../data/2021/reference/exam.xlsx')
rcc1 = util.parse_response(PATH_RCC1)
rcc2 = util.parse_response(PATH_RCC2)
rcc3 = util.parse_response(PATH_RCC3)
rqt1 = util.parse_response(PATH_RQT1)
rthilt1 = util.parse_response(PATH_RTHILT1)
rthilt2 = util.parse_response(PATH_RTHILT2)
rthith1 = util.parse_response(PATH_RTHITH1)
rthith2 = util.parse_response(PATH_RTHITH2)
```

```python
# tinh diem, append cac bai kiem tra
cc1 = mh.get_mark(rcc1, cutpoint, total=15, exam_id='thdc-2021-yrd1-cc1')
cc2 = mh.get_mark(rcc2, cutpoint, total=10, exam_id='thdc-2021-yrd1-cc2')
cc3 = mh.get_mark(rcc3, cutpoint, total=20, exam_id='thdc-2021-yrd1-cc3')
qt1 = mh.get_mark(rqt1, cutpoint, total=30, exam_id='thdc-2021-yrd1-qt1')
thilt1 = mh.get_mark(rthilt1, cutpoint, total=30, exam_id='thdc-2021-yrd1-thilt1')
thilt2 = mh.get_mark(rthilt2, cutpoint, total=30, exam_id='thdc-2021-yrd1-thilt2')
thith1 = mh.get_mark(rthith1, cutpoint, total=30, exam_id='thdc-2021-yrd1-thith1')
thith2 = mh.get_mark(rthith2, cutpoint, total=30, exam_id='thdc-2021-yrd1-thith2')
res = (
    cc1
    .append(cc2)
    .append(cc3)
    .append(qt1)
    .append(thilt1)
    .append(thith1)
    .append(thilt2)
    .append(thith2))
```

```python
def write_1(attendee, mark_long, exam, mon='', lop='', lop_hp='', lan_thi='1', footer='', output='',
             name='', ngay_thi='01/01/2022', ngay='01', thang='01', nam='2022'):
    examinee = util.get_examinee(attendee, mark_long, exam, how='left')
    df = mh.get_mark_thdc_1(attendee, examinee)
    # Mau import diem thdc lan 1, y-2021
    wh.write_mau_import_diem_thdc_1(
        df,
        path=f'../data/2021/mau-import-diem-export/mau-import-diem-{name}-lan{lan_thi}.xlsx',
        mon=mon,
        lop=lop,
        lop_hp=lop_hp,
        lan_thi=lan_thi)

    # Bang diem thdc lan 1, y-2021
    wh.write_bang_diem_thdc_1(
        df,
        path=f'../data/2021/bang-diem/bang-diem-{name}-lan{lan_thi}.xlsx',
        footer=footer,
        mon=mon,
        lop=lop,
        lop_hp=lop_hp,
        lan_thi=lan_thi,
        ngay_thi=ngay_thi,
        ngay=ngay, thang=thang, nam=nam)
```

```python
# thdc-2021-yrd1-y-2021-lan1
write_1(
    attendee11, mark_long, exam, name='thdc-2021-yrd1-y-2021',
    mon='Tin học [2TC]',
    lop='Y-2021',
    lop_hp='211TTO_01000603',
    lan_thi='1',
    ngay_thi='01/01/2022',
    ngay='01', thang='01', nam='2022'
)
```
